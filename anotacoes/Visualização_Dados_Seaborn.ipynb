2.1. - Introdução aos gráficos e subgráficos relacionais

- Descrever a relação entre duas variáveis quantitativas

- Relplot: Gráfico relacional que permite a visualização da relação de duas variáveis quantitativas a partir de um gráfico de dispersão ou gráfico de linhas. Também permite criar subgráficos em uma única figura

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(
  x="total_bill",
  y="tip",
  data=tips,
  kind="scatter",
  col="smoker",
  row="time",
  col+wrap=2,
  col_order=["Thur", "Fri", "Sat", "Sun"] # ou row_order=[]
)
plt.show()
```

- Exercícios:

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3",
            data=student_data,
            kind="scatter",
            row="study_time")

# Show plot
plt.show()
```

3.2. - Criando um box plot

- Um diagrama de caixas mosta a distribuição de dados quantitativos, mostrando a mediana, dispersão, assimetria e outliers, além de facilitar a comparação entre grupos

```python
import matplotlib.pyplot as plt
import seaborn as sns

g = sns.catplot(
  x="time",
  y="total_bill",
  data=tips,
  kind="box",
  order=["Dinner", "Lunch"],
  showfliers=False # remove outliers
)

plt.show()
```

- Por parão os bigodes se estendem até 1.5 * o intervalo interquartil

- Bigodes podem ser estendidos até 2.0 * IQR: whis=2.0

- Mostre os valores mínimos e máximos: whis=[0, 100]

- Exercícios:

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(
  x="internet",
  y="G3",
  data=student_data,
  kind="box",
  hue="location",
  showfliers=False
)


# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(
  x="romantic", y="G3",
  data=student_data,
  kind="box",
  whis=[0, 100]
)

# Show plot
plt.show()
```


4.1. - Como alterar o estilo e a cor do gráfico

- Por que mudar? Preferência pessoal ou melhorar a legibilidade

- A opção "style" da figura inclui o fundo e os eixos

- Opções predefinidas: "white", "dark", "whitegrid", "darkgrid", e "ticks" -> sns.set_style()

- A opção "palette" da figura muda a cor dos principais elementos do gráfico -> sns.set_palette()

- Uso de paletras predefinidas ou personalizadas (RdBu, PRGn, RdBu_r, PRGn_r)

- Paletas customizadas:

```python
custom_palette = ["red", "green", "orange", "blue", "yellow", "purple"]

sns.set_palette(custom_palette)
```

- A opção "context" da figura altera a escala dos elementos e dos rótulos do gráfico -> sns.set_context()

- Do menor para o maior: "paper", "notebook", "talk" e "poster"

- Exercícios:

```python
# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes",
                  "Often", "Always"]

sns.catplot(x="Parents Advice",
            data=survey_data,
            kind="count",
            order=category_order)

# Show plot
plt.show()
```
