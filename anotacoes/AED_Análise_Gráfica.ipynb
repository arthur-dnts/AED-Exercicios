1.4. - Plot all of your data: ECDFs

- Um problema da plotagem de dados utilizando o bee swarm é que, em grandes volumes, os valores podem sobrepor outros, causando a dificuldade da leitura do gráfico

- Para casso com uma maior quantidade de dados o recomendável é utilizar uma função de distribuição cumulativa empírica (Empirical cumulative distribution function - ECDF)

- O eixo X desse gráfico é ordenado, então é necessário utilizar um array NumPy com os dados ordenados

```python
import numpy as np

x = np.sort(df["coluna"])
y = np.arange(1, len(x) + 1 / len(x))
_ = plt.plot(x, y, marker=".", linestyle="none")
_ = plt.xlabel("Label do eixo X")
_ = plt.ylabel("ECDF")
plt.margins(0.02) # Mantêm o dados do plot longe das margens
plt.show()
```

- Exercícios:

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n + 1) / n

    return x, y
```
