1.3. Resumo dos dados

- `.groupby` -> Agrupa por categoria

- `df[["col1", "col2", "col3"]].groupby("coluna").mean()` -> Retorna a média dos dados agrupados (pode-se usar também .sum(), .count(), .min(), .max(), .var(), .std())

- `df[["col1", "col2"]].agg(["mean", "std"])` -> .agg aplica funções de agregação em um dataframe

- `df.agg({"col1": ["mean", "std"], "col2": ["median"]})` -> Aplica o mesmo o mesmo conceito, mas aplicando funções de agregação por coluna

- Colunas nomeaadas com as agregações e gráfico de barras

```python
df.groupby("col").agg(
  mean=("col1", "mean"),
  std=("col2", "std"),
  median=("col3", "median")
)

sns.barplot(data=df, x="col1", y="col2")
plt.show()
```

2.1. - Como lidar com dados ausentes

- Dados ausentes podem afetar as distribuições dos dados (ex: modificação da média ou não representação real da realidade)

2.4. Tratamento de valores discrepantes

- Discrepâncias são valores que ficam muito acima de outros dados

- `print(df["coluna"].describe())` -> Para visualizar e localizar manualmente discrepâncias em uma coluna

- Intervalo intterquartil (IQR): Diferença entre os percentuais de 75 e 25

- `seventy_fifth = df["coluna"].quantile(0.75)` -> Para limites acima de 0.75

- `twenty_fifth = df["coluna"].quantile(0.25)` -> Para limites abaixo de 0.25

- `iqr = seventy_fifth - twenty_fifth` -> Para calcular o IQR

- Limite superior: `upper = seventy_fifth + (1.5 * iqr)`

- Limite inferior: `lower = twenty_fifth - (1.5 * iqr)`

- Criando subconjuntos dos dados:
```python
df[(df["coluna"] < lower | (df["coluna"] > upper))] \
  [["col1", "col2", "col3"]]
```

- Excluir discrepâncias: `no_outliers = df[(df["coluna"] > lower) & (df["coluna"] < upper)]`
