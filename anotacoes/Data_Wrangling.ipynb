7.2. - Tabulating and cross-tabulating data

- A construçãao de tabelas de frequência ou tabulações cruzadas de uma variável por outra variável é bastante comum em análises de dados

- Para construir uma tabela de frequências basta fazer `pd.crosstab(index = df["coluna"], columns = "count")` -> Gera  uma tabela de frequência com a quantidade total de cada elemento da coluna

- O código abaixo gera um DataFraame genérico de carros e suas respectivas cores

```python
import pandas as pd

data = {
    "carro": ["Ford", "Fiat", "Ford", "Renault", "BMW", "Fiat", "Fiat", "BMW"],
    "cor": ["preto", "preto", "vermelho", "branco", "azul", "preto", "preto", "prata"]
}

df = pd.DataFrame(data, columns=["carro", "cor"])

df
```

- Para realizar a contagem da quantidade total de carros por marca basta utilizar `pd.crosstab()`

```python
pd.crosstab(index = df["carro"], columns = "total")
```

- No código anterior a coluna de contagem foi definida genericamente como 'total', mas também seria possível gerar uma tabela de frequência de carros pela cor

```python
pd.crosstab(index = df["carro"], columns = df["cor"], margins=True)
```

7.2.1. - Converting a table of counts to a table of proportions

- Os comandos anteriores constroem tabelas de frequências simples que são úteis apenas em alguns casos de situações específicas, não atendendo aos casso em que é necessário uma tabela de proporção e não de contagem total

- Para diferenciar a quantidade total de linhas e colunas é possível renomar a linha e coluna chamada de 'All' do código anterior para refletir melhor o dado armazenado

```python
tabelas = pd.crosstab(index=df["carro"], columns=df["cor"], margins=True)

tabelas.columns = list(tabelas.columns)[0:-1] + ["linhas_totais"]
tabelas.index = list(tabelas.index)[0:-1] + ["colunas_totais"]

tabelas
```

- `list(df)` ou `df.columns` -> Retornam uma lista de todas as tabela do DataFrame

- `list(df.columns)` -> Converte as colunas do DatFrme para uma lista, permitindo todas as manipulações padrões com listas como substituir itens. O mesmo se aplica para `df.index`

- Para gerar uma tabela de proporção basta dividir o DataFrame de frequência pelo total em cada coluna

```python
tabelas/tabelas.loc["colunas_totais"]
```

- A tabela anterior revela que:
  - Por marca de carro:
    1. Fiat tem a maior participação com 37.5% dos carros
    2. BMW e Fort empatam com 25% dos carros
    3. Renault tem a menor participação com 12.5%

  - Por cor de carro:
    1. Todas as cores têm exatamente 1 carro cada (1.0), ou seja, a distribuição de cores é uniforme

- O procedimento para obter a proporção por linha é um pouco diferente:

```python
tabelas.div(tabelas["linhas_totais"], axis=0)
```

7.3. - Transforming and recording a variable

- No cenário real de análise de dados é comum que a variável trabalhada não seja extamente aquilo que é necessário, nesses casos é comum a conversão de dados de valores contínuos (como idade) em outras variáveis menores (como criança, jovem, adulto, etc).

7.3.1. - Creating a transformed variable

- Transformar uma variável significa que será aplicada uma função matemática simples na variável original a fim de criar um novo dado que descreva melhor o que é necessário e que esteja mais próxima do que se deseja para realizar testes estatísticos


- O exemplo seguinte foi extraído do tópico [7.3.1 do livro Learning Statistics with Python](https://ethanweed.github.io/pythonbook/03.03-pragmatic_matters.html#creating-a-transformed-variable)

> On a scale of 1 (strongly disagree) to 7 (strongly agree), to what extent do you agree with the proposition that “Dinosaurs are awesome”?

```python
dados = [1, 3, 5, 2, 1, 6, 7, 4, 5, 7]
```

- O problema é que existe um senso comum em que o ponto central de uma escala deve ser igual a zero (opinião neutra) e os extremos da esquerda e direita representariam a total discordância ou concordância

- Para modificar os dados da pesquisa, basta criar uma equação simples que subtrai 4 de todos os dados coletados

```python
dados = [x-4 for x in dados]
print(dados)

# O mesmo se aplicaria em um array do NumPy
import numpy as np

arr_dados = np.array([1, 3, 5, 2, 1, 6, 7, 4, 5, 7])
arr_dados = arr_dados - 4
print(arr_dados)
```

- Realizar essa operação é interessante para os casos em que o impacto (ou força) de cada dado ou a direção a que eles tendem

- Para calcular a força de opinião é necessário calcular o valor absuluto do dado central usando a função abs()

```python
arr_dados = np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
arr_dados = arr_dados -4
arr_dados = abs(arr_dados)
arr_dados
```

- Para calcular uma variável que contenha apenas a direção da opinião, basta utilizar o método `numpy.sign()`

- O método sign() do numpy converte todos os números negativos em -1 e todos os números positivos são convertidos para 1 (zeros permancem inalterados).

```python
arr_dados = np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
arr_dados = arr_dados - 4
arr_dados = np.sign(arr_dados)
arr_dados
```

- Com essas três variáveis transformadas é possível montar um DataFrame para analisar as diferenças

```python
import pandas as pd
df = pd.DataFrame(
    {"pontos": np.array([1, 7, 3, 4, 4, 4, 2, 6, 5, 5])
    })

df["centro"] = df["pontos"]-4
df["forca_opiniao"] = abs(df["centro"])
df["direcao_opiniao"] = np.sign(df["pontos"]-4)
df
```

7.3.2. Cutting a numeric variable into categories

- Em alguns casos é necessário cortar uma variável numérica em categorias discretas, por exemplo:

```python
import pandas as pd
df = pd.DataFrame(
    {"idade": np.array([12, 18, 60, 25, 32, 44, 13, 9, 57, 33])
    })

df
```

- Em alguns casos seria mais útil agrupar essas idades em grupos categóricos menores

- jovens (0-20); adultos (21-40); idosos (41-60)

- Pandas possui o método `.cut()` para o agrupamento de dados

```python
df["categorias"] = pd.cut(x = df["idade"], bins = [0,20,40,60], labels = ["jovem", "adulto", "idoso"])

df
```

- No argumento `bins` foram passados 4 valores (0, 20, 40 e 60), mas há somente 3 categorias. Isso é necessário porque o Pandas exige que sejam passados as bordas das categorias ao invés do ponto central

- O método cut possui muitos argumentos, são eles:
  - x: A variável que precisa ser categorizada
  - bins: Um vetor que contêm as localizações dos pontos de separação das categorias ou o número que indica a quantidade de categorias desejadas
  - labels: Os rótulos anexados para cada categoria (opcional)

- O usuário pode deixar que o Python decida por si só como categorizar os dados. O próximo código só irá definir a quantidade de categorias desejadas

```python
df["categorias"] = pd.cut(x = df["idade"], bins = 3)

df
```

- Todas as informação importantes podem ser extraídas olhando os dados tabulados

- O código abaixo irá exibir a quantidade total de itens por categoria

```python
pd.crosstab(index = df["categorias"], columns = "total")
```

- O que o Python faz é tentar dividir as idades em 3 categorias de bins com tamanhos iguais

- Em alguns casos será necessário em categorias de tamanhos diferentes, mas com um número idêntico de casos, para essas situações será necessário adicionar os limites correspondentes em porcentagem

- Utilizando o método `qcut()` que divide por quantidades iguais em percentis, gerando grupos de tamanhos diferentes, mas com quantidades iguais

- Ou seja:
  - `cut()` -> Corta os dados em pedaços de tamanhos fixos
  - `qcut()` -> Corta os dados para que cada grupo tenha o mesmo número de elementos (divide em grupos com quantidade de itens equilibrados)

```python
df["categorias"] = pd.qcut(x = df["idade"], q = [0, .33, .66, 1])

df
```
