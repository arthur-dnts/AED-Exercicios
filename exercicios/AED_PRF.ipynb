```python
# Padronização de todos os datasets antes de prosseguir com as demais questões
import zipfile
import shutil
import os

# Funções úteis

def new_dir(caminho, pastas):
  """
    Cria novos diretórios a partir de um array
    e extrai os arquivos zipados dinamicamente
  """
  try:
    for pasta in pastas:
      # Cria uma nova pasta caso ainda não exista
      if not os.path.isdir(f"{caminho}/{pasta}"):
        os.mkdir(f"{caminho}/{pasta}")

      # Extrai os arquivos dinamicamente
      arquivo = f"{caminho}/{pasta}.zip"
      with zipfile.ZipFile(f"{arquivo}") as dataset:
        dataset.extractall(path=f"{caminho}/{pasta}")
    print("\033[32mArquivos extraídos com sucesso!\033[m")

  except Exception as e:
    print(f"\033[31mErro ao extrair os arquivos: {e}\033[m")

def move_file(caminho, pastas):
  """
    Percorre os diretórios raiz e, se houver
    subpastas, move os arquivos para a pasta
    raiz e remove a subpasta
  """
  try:
    for pasta in pastas:
      pasta_raiz = f"{caminho}/{pasta}"
      # Verifica se há uma subpasta na pasta atual
      subdir = [name for name in os.listdir(pasta_raiz) if os.path.isdir(os.path.join(pasta_raiz, name))]
      if subdir:
        subpasta = f"{pasta_raiz}/{subdir[0]}"
        for arquivo in os.listdir(subpasta):
          if arquivo.endswith(".csv"):
            origem = os.path.join(subpasta, arquivo)
            destino = os.path.join(pasta_raiz, arquivo)
            shutil.move(origem, destino)
        # Remove a subpasta
        shutil.rmtree(subpasta)
    print("\033[32mArquivos movidos com sucesso!\033[m")

  except Exception as e:
    print(f"\033[31mErro ao mover arquivos de subpastas: {e}\033[m")

print("\033[32mImportações e carregamento de funções realizadas com sucesso!\033[m")
```

```python
# Datasets de INFRAÇÕES (MULTAS) 2018, 2021 e 2025
# Criando as pastas e extraindo os arquivos .zip de infrações 2018-2025
pastas = [
    "infracoes_2018", "infracoes_2021", "infracoes_2025"
]

infra_path = "/content/infracoes"

# Cria pastas individuais para cada infração por ano chamando a função new_dir
new_dir(infra_path, pastas)

# Algumas pastas não seguem o padrão de infracoes_20XX_XX.csv
# É necessário realizar o tratamento individual dessas pastas
# para que todos os datasets sigam o mesmo modelo

# Checa se dentro do diretório raiz há uma subpasta
move_file(infra_path, pastas)
```

```python
# Alguns arquivos não seguem o padrão de infracoes_ANO_MES.csv
# esses arquivos precisam ser renomeados para se adequar ao padrão

anos = [
    "2018", "2021", "2025"
]

meses = ["jan", "fev", "mar", "abr", "mai", "jun",
         "jul", "ago", "set", "out", "nov", "dez"]

# Laço para renomear arquivos
for i, pasta in enumerate(pastas):
    caminho_pasta = f"{infra_path}/{pasta}"

    for arquivo in os.listdir(caminho_pasta):
        nome, ext = os.path.splitext(arquivo)

        # Arquivo no padrão correto —> ignora
        if nome.startswith("infracoes_"):
            continue

        # Arquivo no formato infraçoes_ANO_MES.csv
        elif nome.startswith("infraçoes_"):
            novo_nome = nome.replace("infraçoes_", "infracoes_") + ".csv"

        # Arquivo no formato infraçoesANO_MES.csv
        elif nome.startswith("infraçoes"):
            novo_nome = nome.replace("infraçoes", "infracoes_") + ".csv"

        # Arquivo no formato MES.csv
        elif nome in meses:
            mes_num = meses.index(nome) + 1
            novo_nome = f"infracoes_{anos[i]}_{mes_num:02d}.csv"

        os.rename(
            os.path.join(caminho_pasta, arquivo),
            os.path.join(caminho_pasta, novo_nome)
        )

print("\033[32mArquivo renomeados com sucesso!\033[m")
```

**Tarefa 3: Limpeza e preparação dos conjuntos de dados pós-combinação**

```python
# Remoção de células duplicadas
df_aci_final.drop_duplicates()

# Dicionário para o preenchimento de células NaN por coluna
estrategias_nan = {
    "causa_acidente":         "Ignorado",
    "fase_dia":               "Ignorado",
    "condicao_metereologica": "Ignorado",
    "tipo_acidente":          "Ignorado",
    "classificacao_acidente": "Ignorado",
    "sentido_via":            "Ignorado",
    "tracado_via":            "Ignorado",
    "tipo_pista":             "Ignorado",
    "uso_solo":               "Ignorado",
    "mortos":                 0,
    "feridos_leves":          0,
    "feridos_graves":         0,
    "ilesos":                 0,
}

# Aplicando apenas nas colunas que existem no DataFrame
for col, valor in estrategias_nan.items():
    if col in df_aci_final.columns:
        df_aci_final[col] = df_aci_final[col].fillna(valor)

df_aci_final.info()
```

2 - Transformar variáveis para os tipos corretos:
- variáveis relacionadas a tempo (datas, dias, horas) devem ser tratadas como datetimes;
- variáveis numéricas não podem ser trabalhadas como string;
- variáveis categóricas devem ser tratadas como category

```python
# Convertendo colunas com data e hora para datetime
df_aci_final["data_inversa"] = pd.to_datetime(df_aci_final["data_inversa"])
df_aci_final["horario"] = pd.to_datetime(df_aci_final["horario"], format="%H:%M:%S")

# Convertendo células numéricas para float
cols_numericas = ["km", "latitude", "longitude"]

for col in cols_numericas:
    if df_aci_final[col].dtype == object:  # só aplica str.replace se for string
        df_aci_final[col] = df_aci_final[col].str.replace(",", ".", regex=False)
    df_aci_final[col] = pd.to_numeric(df_aci_final[col], errors='coerce')

df_aci_final[cols_numericas].dtypes

# Convertendo colunas categóricas para category
cols_categoricas = [
    "dia_semana", "uf", "causa_acidente", "tipo_acidente",
    "classificacao_acidente", "fase_dia", "sentido_via",
    "condicao_metereologica", "tipo_pista", "tracado_via", "uso_solo"
]

df_aci_final[cols_categoricas] = df_aci_final[cols_categoricas].astype("category")
df_aci_final.info()
df_aci_final.head()
```

3 - Ajustar a codificação das variáveis categóricas, de acordo com o dicionário de cada conjunto,
i.e., caso os dados estajem codificados de alguma forma, transformar os valores codificados nos
valores associados, assinalados o dicionário de dados do conjunto.

```python
# Ao analisar o DataFrame, a coluna uso_solo que reflete
# se o acidente ocorreu em ambiente urbano ou rural utiliza
# apenas 'sim'(urbano) e 'não' (rural), necessitando de tratamento

mapa_uso_solo = {
    "Sim": "Urbano",
    "Não": "Rural",
}

df_aci_final["uso_solo"] = df_aci_final["uso_solo"].map(mapa_uso_solo).astype("category")

# Exibindo a coluna para confirmar a mudança
df_aci_final["uso_solo"].head(10)
```
