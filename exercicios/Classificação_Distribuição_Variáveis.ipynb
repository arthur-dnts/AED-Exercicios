# Tarefa 3: Categorização de Variáveis Numéricas (Binning)

1. A categorização (ou binning) é útil para reduzir o ruído de variáveis contínuas ou criar perfis
demográficos, transformando variáveis quantitativas em qualitativas ordinais.

2. Utilizando a variável quantitativa idade, crie uma nova coluna no seu DataFrame chamada
faixa_etaria. Utilize a seguinte regra para a discretização (intervalos fechados à esquerda e
abertos à direita):

- [0, 18): Menores de idade
- [18, 30): Jovens
- [30, 60): Adultos
- [60, 120): Idosos

```python
# Criação da nova coluna com binning
bins = [0, 18, 30, 60, 120]
labels = ["Menor de Idade", "Jovem", "Adulto", "Idoso"]
df["faixa_etaria"] = pd.cut(df["idade"], bins=bins, labels=labels, right=False)

print(df["idade"].head())
print("\n", df["faixa_etaria"].head())
```

#Tarefa 4: Distribuição de Frequências para Variáveis Categóricas

Após a criação da variável categórica faixa_etaria, precisamos entender o perfil das vítimas
dos acidentes. Suponha que, ao extrair uma amostra do dataset, você obteve a seguinte
contagem para a nova variável:
- Menores de idade: 150 ocorrências
- Jovens: 320 ocorrências
- Adultos: 430 ocorrências
- Idosos: 100 ocorrências

  a) Construa a tabela de distribuição de frequências completa para a variável faixa_etaria.
  
  Sua tabela deve conter as seguintes colunas:

  - Frequência Absoluta (fi)
  - Frequência Relativa (fr) expressa em porcentagem
  - Frequência Acumulada (Fi)
  - Frequência Relativa Acumulada (Fr) expressa em porcentagem

  Um exemplo de como a tabela deve ficar é

  | Faixa Etária | *fi* | *fr* (%) | Fi | Fr (%) |
  | :-------- | :--: | :------ | :-- | :---- |
  | Menores de idade | 150 | 15% | 150 | 15% |
  | Jovens | 320 | 32% | 470 | 47% |
  | Adultos | 430 | 43% | 900 | 90% |
  | Idosos | 100 | 10% | 1000 | 100% |

  b) Com base exclusivamente na tabela gerada no item (a), elabore um parágrafo interpretativo. Responda explicitamente: Qual é o grupo de maior risco de envolvimento em acidentes?

  Qual é a porcentagem total de vítimas que possuem menos de 60 anos? Como essa informação
  de frequência acumulada pode ser útil para políticas públicas de prevenção de acidentes da PRF?
  

```python
# Quantidade total de elementos em faixa_etaria
total = df['faixa_etaria'].value_counts().sum()
tam = len(df['faixa_etaria'].value_counts())

# Lista para armazenamento dos dados
fi = []
fr = []
Fr = []
Fi = []

freqAcu = 0
freqRelAcu = 0
variaveis = ["Menor de Idade", "Jovem", "Adulto", "Idoso"]
for x in range(tam):
  # Calcula a frequência absoluta
  totVar = (df["faixa_etaria"] == f"{variaveis[x]}").sum()
  fi.append(totVar)

  # Calcula a frequência acumulada
  freqAcu += fi[x]

  # Calcula a frequência relativa
  freqRel = (fi[x] * 100) / total

  # Calcula a frequência relativa acumulada
  freqRelAcu += freqRel.round(0)

  # Adiciona os dados nas respectivas listas
  fr.append(f"{freqRel:.0f}%")
  Fr.append(f"{freqRelAcu:.0f}%")
  Fi.append(freqAcu)

# Criação de um DataFrame com os dados
table = {
    "Faixa Etária": variaveis,
    "fi": fi,
    "fr": fr,
    "Fi": Fi,
    "Fr": Fr
}

tabela = pd.DataFrame(table)
tabela.head()
```

A partir da análise da tabela é possível estabelecer que:

- Adultos são o grupo de maior risco em acidentes, tanto em termos absolutos quanto relativos. Pessoas nessa faixa etária ocupam 49% dos acidentes documentados, somando quase 90 mil casos apenas em 2018

- A porcentagem total de vítimas que possuem menos de 60 anos (menores de idade + jovens + adultos) totaliza 92% dos acidentes documentados. Essa informação é importante pois, a partir dela, políticas públicas podem ser tomadas para conscientização e prevenção de acidentes focadas em pessoas abaixo dos 60 anos. Um exemplo seria a conscientização do seguimento das leis de trânsito, treinamento para condutores de veículos e aumento da fiscalização focado em pessoas que se encaixam nessa categoria
